Import Depedencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Data Analysis

```python
# load data
breast_cancer_data = pd.read_csv('data.csv')
```

```python
breast_cancer_data.head()
```

```python
# getting info
breast_cancer_data.info()
```

```python
# replace non numeric data
breast_cancer_data['diagnosis'] = breast_cancer_data['diagnosis'].replace('B', 0)
breast_cancer_data['diagnosis'] = breast_cancer_data['diagnosis'].replace('M', 1)
```

```python
breast_cancer_data.head()
```

```python
# number of row and column
breast_cancer_data.shape
```

```python
# check missing value
breast_cancer_data.isnull().sum()
```

```python
breast_cancer_data.columns
```

```python
# remove column
breast_cancer_data = breast_cancer_data.drop(columns='id', axis=1)
```

```python
# statical measure
breast_cancer_data.describe()
```

```python
# checking distribution
breast_cancer_data['diagnosis'].value_counts()
```

Benign -> 0

Maligant -> 1

```python
breast_cancer_data.groupby('diagnosis').mean()
```

Separating the feature and target

```python
X = breast_cancer_data.drop(columns='diagnosis', axis=1)
y = breast_cancer_data['diagnosis']
```

Standardize Data

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
scaler.transform(X)
```

```python
print(X)
```

```python
print(y)
```

Splitting the data into training and testing data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = make_pipeline(StandardScaler(), LogisticRegression(max_iter=500))
```

```python
# training the logistic regression model
model.fit(X_train, y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy on training data
X_train_pred = model.predict(X_train)
training_data_accuracy = accuracy_score(y_train, X_train_pred)
```

```python
print('Accuracy Score on Training data : ', training_data_accuracy)
```

```python
# accuracy on testing data
X_test_pred = model.predict(X_test)
test_data_accuracy = accuracy_score(y_test, X_test_pred)
```

```python
print('Accuracy Score on Testing data : ', test_data_accuracy)
```

Building a Predictive System

```python
input_data = (13,21.82,87.5,519.8,0.1273,0.1932,0.1859,0.09353,0.235,0.07389,0.3063,1.002,2.406,24.32,0.005731,0.03502,0.03553,0.01226,0.02143,0.003749,15.49,30.73,106.2,739.3,0.1703,0.5401,0.539,0.206,0.4378,0.1072)

# change input data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the array
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

# standardize the input data
std_data = scaler.transform(input_data_reshaped)

prediction = model.predict(std_data)
print(prediction)

if (prediction[0] == 0):
  print('The breast cancer is Benign')

else:
  print('The breast cancer is Malignant')
```
