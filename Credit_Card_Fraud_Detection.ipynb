Import Depedencies

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

Data Collection and Analysis

```python
# load data
credit_card_data = pd.read_csv('creditcard.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
credit_card_data.shape
```

```python
# getting info
credit_card_data.info()
```

```python
# checking missing values
credit_card_data.isnull().sum()
```

```python
# handling missing value
credit_card_data = credit_card_data.dropna()
```

```python
credit_card_data.isnull().sum()
```

```python
# distribution of legit transaction and fraudulent transaction
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x='Class', data=credit_card_data)
plt.title('Class Distribution')
plt.show()
```

Imbalance Class

0 -> normal
1 -> fraud

```python
# separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit)
```

```python
print(fraud)
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# getting measure statistical
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the value
credit_card_data.groupby('Class').mean()
```

Under Sampling

Build a sample dataset containing similar distribution of normal transaction and fraud transaction

Number of fraud transaction -> 38

```python
legit_sample = legit.sample(n=38)
```

Concat two dataframe

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

Splitting the data into feature and target

```python
X = new_dataset.drop(columns='Class', axis=1)
y = new_dataset['Class']
```

```python
print(X)
```

```python
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model training

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

Model Evaluation

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, y_train)
```

```python
print('Accuracy Score of Training data : ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, y_test)
```

```python
print('Accuracy Score of Test data : ', test_data_accuracy)
```

```python
# confussion matrix
cm = confusion_matrix(y_test, X_test_prediction)

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
